## 创建结点

```python
import networkx as nx
import numpy as np
import random
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib as mpl
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
G = nx.Graph()
```

```python
nx.draw(G)
```

```python
G.add_node('刘备')
```

```python
G.nodes
```

```python
G.add_node('Tommy')
```

```python
G.nodes
```

```python
G.add_node(1)
```

```python
G.nodes
```

```python
G.add_nodes_from(['诸葛亮', '曹操'])
```

```python
# 添加带属性的结点
G.add_nodes_from([
    ('关羽', {'武器':'青龙偃月刀'}), 
    ('张飞', {'武器':'丈八蛇矛'})
])
```

```python
G.nodes
```

```python
nx.draw(G)
```

```python
G.nodes(data=True)
```

```python
for node in G.nodes(data=True):
    print(node)
```

## 创建连接

```python
G.add_edge('刘备', '诸葛亮', weight=0.5, like=3)
```

```python
nx.draw(G, with_labels=True)
```

```python
G.nodes['张飞']
```
